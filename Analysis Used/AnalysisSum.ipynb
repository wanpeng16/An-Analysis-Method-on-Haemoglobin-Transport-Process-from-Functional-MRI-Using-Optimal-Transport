```python
def classstas(Class):
    fMap =0
    iMap =0
    for i in range(len(Class)):
        fdirc = Class[i]['fdir']
        idirc = Class[i]['idir']
        fm = np.load(fdirc)
        im = np.load(idirc)
        fm = fm/fm.sum()
        fMap = fMap + fm
        iMap = iMap + im
    return 500*(fMap/fMap.sum()), 500*(iMap/iMap.sum())
```

```python
def plotbar(data, name):
    plt.figure(figsize=(5, 5))
    groups_len = len(data)
    fig, ax = plt.subplots()
    index = np.arange(groups_len)
    opacity = 0.9
    rects1 = ax.bar(index, data,
                alpha=opacity, color='b',
                label=name)
    
    ax.set_xlabel('Voxels')
    ax.set_ylabel('Comunication Level')
    ax.set_title(name)
    ax.legend()
    fig.tight_layout()
    plt.savefig("%s.pgf"%name)
    plt.show()
```

```python
def Dplotbar(data, name,data2, name2):
    groups_len = len(data)
    fig, ax = plt.subplots()
    index = np.arange(groups_len)
    opacity = 0.4
    rects1 = ax.bar(index, data,
                alpha=opacity, color='b',
                label=name)
    rects2 = ax.bar(index, data2,
                alpha=opacity, color='r',
                label=name2)
    ax.set_xlabel('Voxels')
    ax.set_ylabel('Comunication Level')
    #ax.set_title(namename)
    ax.legend()
    fig.tight_layout()
    plt.axis('tight')
    plt.figure(figsize=(5, 5))
    plt.savefig('Flux.') 
    plt.show()
```

```python
def segform(data):
    x, y, z = data.shape
    segdata = np.zeros(data.shape)
    statistic = np.zeros(data.max())
    for i in range(x):
        for j in range(y):
            for k in range(z):
                if data[i,j,k] >0:
                    statistic[data[i,j,k]-1] = statistic[data[i,j,k]-1] + 1
    
    return statistic
```
