```python
import numpy as np
import matplotlib.pyplot as plt
import os
from nilearn import plotting
```

```python
root = "C:\\Users\Yin\Desktop\ADHD200\OTMcorelation"
dirc = os.listdir(root)
```

```python
Mdirc = os.path.join(root,dirc[1])
Ndirc = os.path.join(root,dirc[2])
```

```python
CorelationMap = np.load(Mdirc)
Names = np.load(Ndirc)
```

```python
for i in range(len(Names)):
    Names[i] = Names[i].replace('stasMap', ' ')
    Names[i] = Names[i].replace('.npy', ' ')
    Names[i] = Names[i].replace('_rest_tshift_RPI_voreg_mni.nii.gz.npy',' ')
```

```python
plt.figure()
pic = plotting.plot_matrix(CorelationMap, labels=Names, colorbar=True)
plt.show()
```
