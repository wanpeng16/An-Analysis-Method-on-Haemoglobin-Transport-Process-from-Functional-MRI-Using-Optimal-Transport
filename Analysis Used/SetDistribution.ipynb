```python
Map1000mid = 300
Map1000mim = 10
Map1000max = 1000
```

```python
def setdistribution(Mdic, mid, mim, Max):
    M = np.load(Mdic)
    M = M/M.sum()
    X = M - M.T
    ef = 1000/(abs(X).max())
    Map1000 = ef*X
    Map1000mid = mid
    Map1000mim = mim
    Map1000max = Max
    d1 = (Map1000mid - Map1000mim)/200
    para = (55*d1)/(Map1000max-Map1000mid)
    TM = np.zeros(8468*8468)
    TM = TM.reshape(8468,8468)
    for i in range(8468):
        for j in range(i + 1, 8468):
            TM[i,j] = Map1000[i,j]
    for i in range(8468):
        for j in range(8468):
            if TM[i,j] > 0:
                if TM[i,j] <Map1000mim:
                    TM[i,j]=0
                if TM[i,j] >Map1000mid:
                    TM[i,j] = para*(TM[i,j]-Map1000mid)+Map1000mid
            if TM[i,j] < 0:
                v = -TM[i,j]
                if v <Map1000mim:
                    TM[i,j]=0
                if v >Map1000mid:
                    TM[i,j] = -(para*(v-Map1000mid)+Map1000mid)
    et = 200/Map1000mid
    HotM = TM*et
    return HotM
```

```python
HotMNormalUp15 = setdistribution(M1dic,mid = 300, mim = 10, Max = 1000)
```
